```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = ("Python DataScience Education Research Food Travel Business Photography Gadgets Love Study Adventure Books NLP Water Tech Programming Sleep Work Nature Pictures Friendship")
```

```python
wordcloud = WordCloud(width=1263, height=370,  background_color="white", margin=0).generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
